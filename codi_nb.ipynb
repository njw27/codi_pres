## Now we can do some insights for total contract value.


We use several financial metrics at Codi to evaluate the performance of our business, one of which is total contract value. Put simply, the TCV represents the full value of a customers contract across an agreed-upon term, inclusive of all recurring revenue and one-time fees. We have two types of contracts: recurring (known internally as long-term plans or LTPs) and nonrecurring (known as short-term bookings or STBs).  Note: The date a contract is signed is used to determine which month to attribute LTP for the first term and subsequent terms use the renewed contract start date, whereas the start date is always used for STBs.) We also include add-on revenue in our TCV calculation, which can be either one-time or recurring.

```python
ytd_l = long_w_names[["name","amount_monthly","payment_interval_months","term_length_months","renews_automatically",\
                      "dt_contract_signed_date"]]    # need to calc start month total

ytd_l["recurring_payment"] = ytd_l.amount_monthly*ytd_l.payment_interval_months


# Start
# #look at term months total
# time counter = 0
# start with signed date as date -> date
# pay monthly *interval 
# add interval len to time counter
# check if time counter = term months
# if greater you messed up
# if equal reset counter to 0 increae pay monthly and reloop if renews autmatically is true
# if less add months of time counter to dt months (will add up automatically)
# entry at this new date with pay monthly *interval 
# add interval len to time counter
# repeat
# when year >2024 end 


date_series = []

name_series = []

amount_series= []

def thing(renewal, date, term, name, amount, interval):
    counter = 0
    while counter <= term-interval:
        date_series.append(date)
        name_series.append(name)
        amount_series.append(amount*interval)
        counter+= interval
        date+= relativedelta(months=interval)
        if renewal == True and date < pd.Timestamp('2025-01-01') and counter == term:
            counter = 0
            amount = amount *1.03 #3% term increase
        else:
            pass
        
        
        
            

    



long_monthly = long_w_names.apply(lambda x: thing(renewal = x.renews_automatically, date = x.dt_contract_signed_date,\
                                                  term = x.term_length_months, name = x["name"], \
                                                  amount=x.amount_monthly, interval= x.payment_interval_months), \
                                  axis=1)



framed_long = pd.DataFrame(
    {'date': date_series,
     'amount': amount_series,
     'name': name_series
    }) 


framed_long["returning_cust"] = 0
framed_long["bookings"] = 1
framed_long["month"] = pd.DatetimeIndex(framed_long["date"]).month
framed_long["year"] = pd.DatetimeIndex(framed_long["date"]).year

framed_long = framed_long.sort_values(by = ["year", "month"])
framed_long = framed_long.reset_index()
framed_long = framed_long [["name", "returning_cust", "bookings", "month", "year", "amount"]]
framed_long["returning_cust"] = framed_long["name"].duplicated()

framed_long.drop(framed_long.loc[framed_long['year']==2025].index, inplace=True)

framed_long
# long_monthly_no_addons =  = ytd_l[]
# long_monthly_w_addons = ytd_l[]
```
